# Combine

```python
import pandas as pd
import numpy as np

for fold in [0,1,2,3,4]:
    print("*"*50)
    print("FOLD = ", fold)
    print("*"*50)
    pred_df = None
    for CLASS in [
        'cohesion',
        'syntax','vocabulary','phraseology','grammar','conventions'
    ]:
        print("*"*50)
        print(CLASS)
        print("*"*50)
        tmp_df = pd.read_csv(f'./result/{VERSION}/pseudo_{CLASS}_fold{fold}.csv')
        if pred_df is None:
            pred_df = tmp_df.copy()
        else:
            pred_df = pred_df.merge(tmp_df, on="text_id", how="left")

    test_df = pd.read_csv(UNLABELED_DATA_PATH)
    test_df = test_df.rename(columns={'id':'text_id'})
    pseudo_df = test_df[['text_id']].merge(pred_df, on='text_id', how='left')
    pseudo_df.to_csv(f'./result/{VERSION}/pseudo_fold{fold}.csv', index=False)
```
