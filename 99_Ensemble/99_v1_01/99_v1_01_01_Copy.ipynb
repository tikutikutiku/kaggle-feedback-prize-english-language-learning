```python
TARGET_COLS = ['cohesion','syntax','vocabulary','phraseology','grammar','conventions']

preds = oof_df_1[TARGET_COLS].values
preds += oof_df_2[TARGET_COLS].values
preds /= num_models
```

```python
oof_df = oof_df_1.copy()
for i,col in enumerate(TARGET_COLS):
    oof_df[col] = preds[:,i]
```

```python
import numpy as np

def calc_metric(pred, gt):
    '''
    pred : (num_data, num_labels)
    gt : (num_data, num_labels)
    '''
    score = np.sqrt(np.mean((pred - gt)**2, axis=0))
    score = score.mean()
    return score
```

```python
score = calc_metric(pred=oof_df[TARGET_COLS].values, gt=train_df[TARGET_COLS].values)
print('CV={:.5f}'.format(score))

# add clip(1,5)
score = calc_metric(pred=oof_df[TARGET_COLS].values.clip(1,5), gt=train_df[TARGET_COLS].values)
print('CV={:.5f}'.format(score))
```

```python
oofs = []
oofs.append(oof_df_1[TARGET_COLS].values)
oofs.append(oof_df_2[TARGET_COLS].values)
```

```python
ensemble_predictions=np.stack(oofs)
```

```python
def ensemble_score(weights,return_pred=False, add_clip=True):
    weights=np.array(weights)
    weights=weights.reshape(-1,1,1)/weights.sum()
    p=weights.reshape(-1,1,1)*ensemble_predictions
    p=p.sum(0)
    if add_clip:
        p = p.clip(0.5)
        score=calc_metric(p, train_df[TARGET_COLS])
    else:
        score=calc_metric(p, train_df[TARGET_COLS])
    if return_pred:
        return score,p
    else:
        return score   
```

```python
from skopt import gp_minimize

results=gp_minimize(ensemble_score, np.array([[0.1,1] for i in range(len(ensemble_predictions))]),
                    verbose=True,random_state=2022)
```

```python
best_weights=np.array(results['x'])/sum(results['x'])
print(best_weights)
```
